# *Importing required packages/modules*

```
import numpy as np
import matplotlib.pyplot as plt
import math
```

# *Preparing the X-axis values for plotting the equations*

```
#Given that:-
n = 7
k = 4
```

```
X = [i for i in range(1, n+1)]
```

# *Defining Y-axis values for each of the target equations to be plotted*

```
Y_for_logarithmic = [math.log(i, 2) for i in X] #Base of log = 2

Y_for_linear = X

Y_for_quadratic = [i**2 for i in X]

Y_for_cubic = [i**3 for i in X]

Y_for_polynomial = [i**k for i in X]
```

# *Creating the plot*

```
f, (ax, ax2) = plt.subplots(2, 1, figsize=(15, 15), sharex=True)
f.subplots_adjust(hspace=0.05)

f.suptitle('Assignment 1: Visualization of different time complexity equations of algorithms', fontsize = 20)

ax.plot(X,Y_for_cubic,label='$θ(n^{3})$', color = 'red', marker = '*')
ax.plot(X,Y_for_polynomial,label='$θ(n^{k})$', color = 'purple', marker = 's')

ax2.plot(X,Y_for_logarithmic,label='$θ(log (n))$', marker = 'o')
ax2.plot(X,Y_for_linear,label='$θ(n)$', color = 'gold', marker = 'D')
ax2.plot(X,Y_for_quadratic,label='$θ(n^{2})$', color = 'green', marker = 'h')
ax2.plot(X,Y_for_cubic,label='$θ(n^{3})$', color = 'red', marker = '*')
ax2.plot(X,Y_for_polynomial,label='$θ(n^{k})$', color = 'purple', marker = 's')

ax.set_ylim(7.1, 2500)  # cubic and polynomial
ax2.set_ylim(0, 7)  # all five

# hiding the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop='off')  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

d = .015  # how big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax.transAxes,color = 'k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switching to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

ax.grid(True, alpha = 0.5)
ax2.grid(True, alpha = 0.5)
plt.xlabel("<----  n  ---->", fontsize = 15)
plt.ylabel("<----  θ  ---->", fontsize = 15)
ax.legend(loc="upper center", fontsize = 15)
ax2.legend(loc="upper center", fontsize = 12)

# Showing the plot
plt.show()
```
